```python
import pandas as pd
excel_file = 'synonyms.xlsx'
df = pd.read_excel(excel_file)
```

```python
def_new = []
for index, row in df.iterrows():
    defstr = ""
    l = []
    l.append(row['Word'])
    syn1 = str(row['Syn1']).split(';')
##    print (len(syn1[0].split(" ")))
    if(len(syn1)!=0):
        for i in syn1:
            if(i!= 'nan' and len(i.split(" "))>3):
                defstr+=(str(i))
                defstr+=", "
    syn2 = str(row['Syn2']).split(';')
    if(len(syn2)!=0):
        for i in syn2:
            if(i!= 'nan' and len(i.split(" "))>3):
                defstr+=(str(i))
                defstr+=", "
    syn3 = str(row['Syn3']).split(';')
    if(len(syn3)!=0):
        for i in syn3:
            if(i!= 'nan' and len(i.split(" "))>3):
                defstr+=(str(i))
                defstr+=", "
    defstr = defstr[0:-2]
    print (row['Word'], defstr)
    l.append(defstr)
    if(len(l)==2 and len(defstr)>0):
        def_new.append(l)
print (def_new[:10])
##   print(row['Word'], row['POS'])
```

```python
# import xlsxwriter module 
import xlsxwriter 
  
workbook = xlsxwriter.Workbook('def_new_col_1.xlsx', options={'nan_inf_to_errors': True}) 
worksheet = workbook.add_worksheet() 

worksheet.write(0, 0, "Word")
worksheet.write(0, 1, "Def") 

row = 1
# iterating through file list
for i in range(len(def_new)):
    column = 0
    for j in range(len(def_new[i])):
       # write operation perform 
        worksheet.write(row, column, def_new[i][j]) 
        column += 1
        # incrementing the value of row by one 
        # with each iteratons. 
    row += 1 

workbook.close() 
```
